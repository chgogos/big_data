```python
# 1. Create a spark session

from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
spark.version
```

```python
# 2. Count votes by candidate

ballots = [
    "ΚΩΣΤΑΣ", "ΑΝΔΡΕΑΣ", "ΑΝΔΡΕΑΣ", "ΚΩΣΤΑΣ",
    "ΑΝΔΡΕΑΣ", "ΑΝΔΡΕΑΣ", "ΚΩΣΤΑΣ"
]

rdd = spark.sparkContext.parallelize(ballots)

# MAP: emit (candidate, 1) for each vote
mapped = rdd.map(lambda candidate: (candidate, 1))

# REDUCE: sum counts per candidate
vote_counts = mapped.reduceByKey(lambda a, b: a + b)

vote_counts.collect()
```

```python
# 3. word count

lines = [
    "spark makes big data processing simple",
    "spark uses map and reduce",
    "big data needs distributed processing"
]

rdd = spark.sparkContext.parallelize(lines)

word_counts = (
    rdd
    .flatMap(lambda line: line.split())
    # Emit (word, 1)
    .map(lambda word: (word.lower(), 1))
    # Sum counts per word
    .reduceByKey(lambda a, b: a + b)
)

word_counts.collect()
```
