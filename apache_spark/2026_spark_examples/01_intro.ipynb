```python
# Create a spark session
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
spark.version
```

```python
# Create a dataframe from a python list (and a schema)

from datetime import date

students_schema = 'id long, name string, enrolment_date date, gpa double'
students_data = [(1, "Nikos Zikos", date(2025, 9, 1), 7.7),
                 (2, "Maria Pappa", date(2025, 9, 2), 6.7),
                 (2, "Petros Kokkinos", date(2025, 9, 1), 7.6)]
students_df = spark.createDataFrame(students_data, students_schema)
students_df.show()
```

```python
# write dataframe as table "students" using the default format (parquet)
students_df.write.mode("overwrite").saveAsTable("students")
```

```python
# write dataframe as table "students" using orc format
(students_df
    .write
    .format("orc")
    .mode("overwrite")
    .saveAsTable("students_orc")
)
```

```python
# load dataframe from table "students" 
students_df2 = spark.table("students")
students_df2.show()
```

```python
# load dataframe from table "students" using SQL
students_df3 = spark.sql("select * from students order by gpa desc")
students_df3.show()
```

```python
# load dataframe from table "students_orc"
students_df4 = spark.table("students_orc")
students_df4.show()
```
